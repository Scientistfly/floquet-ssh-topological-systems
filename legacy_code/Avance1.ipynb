```python
import numpy as np
```

```python
sigma_x = np.array(((0, 1), (1, 0)))
sigma_y = np.array(((0, -1j),(1j,0)))
```

```python
N=3
M=3
k = -0.3
P = (2*40) + (2*160)
eta = 0.5
r = 160.0/(160.0+40.0)
sp.constants.hbar
```

```python
H = np.zeros(((N*2 + 1)*2, (N*2 + 1)*2), dtype = complex)
for n_f in range(-N, N+1):
    for n_c in range(-N, N+1):
        if(n_f == n_c):
            H[((n_c+N)*2):((n_c+N)*2)+2,((n_f+N)*2):((n_f+N)*2)+2] = (sigma_x*k) + (sigma_y*((2*np.pi)/P)*(n_f-0.5+eta))
        for m in range(-M, M+1):
            if(m != 0):
                if((m % 2) != 0):
                    if(n_f == n_c + m):
                        H[((n_f+N)*2):((n_f+N)*2)+2,((n_c+N)*2):((n_c+N)*2)+2] = sigma_x*sp.constants.e*2*1e-9*1e-9*P*np.power(-1,(m+1.0)/2.0)*(np.sin(m*np.pi*r/2)/(np.power(m,2)*np.power(np.pi,2)*sp.constants.hbar))
H
```

```python
from scipy.sparse.linalg import eigsh
eigenvalues, eigenvectors = eigsh(H, k=6)
eigenvalues*330
```
